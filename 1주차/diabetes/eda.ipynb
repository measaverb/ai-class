# EDA: diabetes.csv

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

## Data Loading

```python
df = pd.read_csv("data/diabetes.csv")
```

## Basic Information

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

## Correlation Matrix

```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(15, 10))
df.boxplot()
plt.title("Box Plot of Features")
plt.xticks(rotation=45)
plt.show()
```

```python
df.hist(figsize=(15, 10), bins=30)
plt.tight_layout()
plt.show()
```

```python
features_to_plot = ['Glucose', 'BMI', 'Age']
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, feature in enumerate(features_to_plot):
    sns.scatterplot(data=df, x=feature, y='Outcome', ax=axes[i])
plt.tight_layout()
plt.show()
```

```python
positive_outcome_percentage = (df['Outcome'].sum() / len(df)) * 100
print(f"Percentage of positive outcomes: {positive_outcome_percentage:.2f}%")
```
