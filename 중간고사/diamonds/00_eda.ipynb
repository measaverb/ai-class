```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('data/diamonds.csv')
```

```python
print(data.info())
print(data.head())
print(data.shape)
```

```python
print(data.describe())
```

```python
print(data.isnull().sum())
```

```python
numerical_features = ['carat', 'depth', 'table', 
                      'x', 'y', 'z']

fig, axes = plt.subplots(3, 3, figsize=(20, 20))
axes = axes.flatten()

for i, feature in enumerate(numerical_features + ['price']):
    sns.boxplot(x='price', y=feature, data=data, ax=axes[i])
    axes[i].set_title(f'{feature} by Price')

plt.tight_layout()
plt.show()
```
